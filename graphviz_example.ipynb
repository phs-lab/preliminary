# Graphviz

- 설치 : https://pypi.org/project/graphviz/ # cf: conda install graphviz python-graphviz -c anaconda
- Graphviz [Pocket Reference](https://graphs.grevian.org/example), [graphviz attribute](https://graphviz.org/doc/info/attrs.html), [User guide](https://graphviz.readthedocs.io/en/stable/manual.html)

```python
src = """
graph simple_graph {
    label = "Pocket Reference - Example 1: Simple Graph";
    labelloc = t;   // t: top, b: bottom
    rankdir = LR;   // LR: Left to Right, RL: Right to Left
    a -- b; b -- c; // TB: Top to Bottom ( = UD: Up Down), BT: Bottom to Top
    a -- c; c -- d;
    a -- e; e -- c;
}"""
my.view_graphviz(src)
```

```python
src = """
digraph simple_digraph {
    label = "Pocket Reference - Example: Simple Digraph";
    labelloc = t;   // t: top, b: bottom
    rankdir = LR;   // LR: Left to Right, RL: Right to Left
    a -> b -> c;    // TB: Top to Bottom ( = UD: Up Down), BT: Bottom to Top
    a -> e -> c;
    a -> c -> d;
}"""
my.view_graphviz(src)
```

```python
src = """
graph {
    rankdir = LR;
    label = "Pocket Reference - Example 5: Showing A Path";
    c -- f [color=red,penwidth=3.0];
    f-- e -- d; 
    c -- b;
    b -- a [color=red,penwidth=3.0];
    a -- d;
    b -- d [color=red,penwidth=3.0];
    c -- d [color=red,penwidth=3.0];
}
"""
my.view_graphviz(src)
```

```python
src = """
graph ni {  // Polygon-based node shapes : https://graphviz.org/doc/info/shapes.html#polygon
    rankdir=LR; 
    label = "User Guide > Attributes"; labelloc = t;
    
    node [shape=rarrow];
    1 [label="Ni!"];
    2 [label="Ni!"];
    3 [label="Ni!" shape=egg];
    node [shape=star];
    4 [label="Ni!"];
    5 [label="Ni!"];
    1 -- 2 -- 3 -- 4 -- 5;
}
"""
my.view_graphviz(src)
```

```python
src = """
digraph structs {  // Record-based node shapes : https://graphviz.org/doc/info/shapes.html#record
    rankdir = TB;
    label = "Graphviz Documentation > shapes > record-based node"; labelloc = t;
    node [shape=record];
    struct1 [label="<f0> left|<f1> mid&#92; dle|<f2> right"];
    struct2 [label="<f0> one|<f1> two"];
    struct3 [label="hello&#92;nworld |{ b |{c|<here> d|e}| f}| g | h"];
    struct1:f1 -> struct2:f0;
    struct1:f2 -> struct3:here;
}
"""
my.view_graphviz(src)
```
